```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = {
    'days': ['mon','tue','wed','thur','fri'],
    'hours': [2,3,2.5,4,4.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,75,80,75],
    'Sleep_hours':[6,7,7,8,8],
    'Assignments_completed':[1,2,1,3,2]
}
df = pd.DataFrame(data)
df
```

```python
# Line Chart (Trend)
plt.plot(df['days'],df['hours'],marker = 'o')
plt.title('Study hours')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.show()
```

```python
# Advanced Line Chart
plt.figure(figsize= (7,4))

plt.plot(df['days'], df['hours'],
         marker = '*', color = 'red', label = 'Marks')

plt.title('Study hours and marks trend')
plt.xlabel('Days')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Bar Chart (Comparison)

plt.bar(df['days'],df['Marks'],color = 'skyblue')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
# Advanced Bar Graph
bars = plt.bar(df['days'],df['Marks'],color = 'green')
for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2, y, str(y),
  ha = 'center',va = 'bottom')

plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')

plt.grid(axis = 'y')
plt.show()
```

```python
# Histogram (Distribution)
# Shows how values are spread

data = {
    'marks': [50,60,55,70,65,84,87,85,83,82,90,68,46,35,78,95,74,84,96],
}
df = pd.DataFrame(data)
plt.hist(df['marks'],bins = 8,edgecolor = 'red', alpha= 1)
plt.title('Marks Histogram')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = {
    'days': ['mon','tue','wed','thur','fri'],
    'hours': [2,3,2.5,4,4.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,75,80,75],
    'Sleep_hours':[6,7,7,8,8],
    'Assignments_completed':[1,2,1,3,2]
}
df = pd.DataFrame(data)
df
```

```python
# Scatter Plot (Relationship)
import matplotlib.pyplot as plt

plt.scatter(df['hours'],df['Marks'],color = 'green')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
```

```python
# Conditional Scatter Plot
# Example category
df['result'] = ['Fail','Pass','Fail','Pass','Pass']
colors = ['red' if result == 'Fail' else 'green' for result in df['result']]
plt.scatter(df['hours'],df['Marks'], c = colors)


plt.title = 'Study Hours vs Marks'
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```
